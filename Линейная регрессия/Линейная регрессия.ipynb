# Линейная регрессия

Особенности:
- Предполагает линейную зависимость (линейную в коэффициентах $\beta_i$) между независимыми переменными X и зависимой переменной y.
- Делает несколько предположений о природе данных.
- Регрессия не является основанием для вывода о существования однозначной причинно-следственной связи между регрессорами и реграссандой*. A regression model does not imply a cause-and-effect relationship between the variables. Even though a strong empirical relationship may exist between two or more variables, this cannot be considered evidence that the regressor variables and the response are related in a cause-and-effect manner. To establish causality, the relationship between the regressors and the response must have a basis outside the sample data—for example, the relationship may be suggested by theoretical considerations. Regression analysis can aid in confirming a cause-and-effect relationship, but it cannot be the sole basis of such a claim.


# 0. Введение

* <font color='brown'>Вопрос:</font> откуда идёт название "регрессия"?
* <font color='brown'>Ответ:</font> тут подразумевается регрессия к среднему - regression to the mean. Фрэнсис Гальтон, генетик и статистик, изучая растение душистый горошек (sweet pea) заметил, что потомки "экстремальных" - то есть очень низких или очень высоких растений - получаются по росту не такими, как их родители, а более средними, более mediocre. Их рост регрессирует от экстремальных значений их родителей обратно к среднему значению роста популяции растений. Подробнее смотри [источник](https://priceonomics.com/the-discovery-of-statistical-regression/).


Представим ситуацию: имеются данные, которые сгенерированы некоей прямой линией (в случае с более чем одним атрибутом X - гиперплоскостью), с добавленным шумом. Например, мы измеряли координату автомобиля при равномерном движении (а это линейная зависимость), но из-за погрешности GPS в данных присутствует некий случайный шум, который мы не можем спрогнозировать. <br>
*Примечание:* разумеется, в ML шум - это не только и не столько погрешность измерений, сколько "сложность модели, которую не могут описать имеющиеся в наличии данные" (complexity uncaptured by data). Например, в магазине продают пиво согласно двум признакам - у покупателя с собой должна быть нужная сумма и покупателю должно быть не меньше 18 лет. Если нам предоставлен только один признак, например - "имеет ли покупатель с собой нужную сумму", то в нашей модели будет присутствовать заметный шум, поскольку для несовершеннолетних покупателей, имеющих с собой нужную сумму, ML-алгоритм выдаст неправильный прогноз, ведь в данных просто не отражён тот важный момент, что покупатель должен быть совершеннолетним.

Итак, имеются данные, сгенерированные некоей прямой или гиперплоскостью. Для простоты иллюстрирования возьмём прямую. Прямая имеет некое смещение $\beta_0$ и некий наклон $\beta_1$: это "истинные" смещение и наклон той самой "истинной" прямой, которая и породила данные. Математическая модель в этом случае записывается как:

$$y(x) = \underbrace{f(x)}_{сигнал} + \underbrace{\varepsilon(x)}_{шум} = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_kx_k + \varepsilon(x) = \sum_{i=1}^{k} \beta_ix_i + \varepsilon(x)$$

где $f(x)$ - это *сигнал*, $\varepsilon(x)$ - это *шум* или ненаблюдаемые ошибки (error term).

Начнём рассматривать пример. Создадим прямую, которая порождает данные с определённым шумом:

```python
# General:
import numpy as np
# Graphics:
import seaborn as sns
from matplotlib import pyplot as plt
# ML:
import statsmodels.api as sm
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def describe_plot(plot, title, xlabel, ylabel):
    plot.title(title)
    plot.xlabel(xlabel)
    plot.ylabel(ylabel)

FONT_BOLD, FONT_END = '\033[1m', '\033[0m'

BIAS  = 4.5
NOISE = 12
POPULATION_SIZE = 500
SAMPLE_SIZE     = int(POPULATION_SIZE/5)

# Сгенерируем более-менее гомоскедастичные данные, чтобы оценка стандартной ошибки не была смещена:
X, y, real_coef = make_regression(n_samples=POPULATION_SIZE, n_features=1, coef=True, noise=NOISE, bias=BIAS)

# Отобразим данные и "истинную линию регрессии":
print('Значения коэффициентов "истинной" регрессионной линии (из make_regression):')
print('\tβ0 =', BIAS, '\n\tβ1 =', real_coef.item(0))
# ---
xx = np.linspace(min(X), max(X), 2)
plt.plot(xx, xx*real_coef.item(0) + BIAS, c='r')
plt.scatter(X, y, marker='.', c='c')
describe_plot(plt, 
              'Сгенерированные данные и "истинная" линия регрессии\nОбъём популяции n = ' + str(POPULATION_SIZE), 
              'x', 'y')
plt.grid(color='b', linestyle='-', linewidth=0.2)
plt.show()

# Вытянем выборку объёмом n и отобразим её:
X_sample, _, y_sample, _ = train_test_split(X, y, test_size = POPULATION_SIZE-SAMPLE_SIZE)
plt.scatter(X_sample, y_sample, marker='.', c='g')
describe_plot(plt, 'Имеющаяся у нас в наличии выборка объёмом n = ' + str(SAMPLE_SIZE), 'x', 'y')
plt.grid(color='b', linestyle='-', linewidth=0.2)
plt.show()
```

На верхнем рисунке - вся генеральная совокупность и "истинная" линия регрессия, порождающая данную совокупность. На нижнем рисунке - та картина, которую в реальности наблюдаем мы.

Это базовое предположение линейной регресии. Наша задача сводится к тому, чтобы на основании имеющихся у нас дата-точек как можно точнее угадать "истинные" коэффициенты $\beta_0$ и $\beta_1$.

><font color='red'>**Предположение 1:**</font> между независимыми переменными и зависимой переменной существует линейная (в коэффициентах $\beta_i$) зависимость.<br>
<font color='red'> Важность: критично</font>
<font color='red'>Как проверить: </font> напрямую - никак. Мы можем только предполагать. Косвенно: используя Residual Plots. <br>

Покажет худший результат, чем другие модели, более точно отражающие зависимости y от X. Кроме того, в этом случае невозможно будет показать, что оценки коэффициентов регрессии несмещённые, да и вообще что они имеют какой-либо смысл (зависимость-то нелинейная, а мы считаем оценки линейных коэффициентов - всё равно что искать коэффициент $y=kx$ для целевой функции $y=2\pi cos(x)$).<br>
<font color='red'>Как исправить нарушение:</font> напрямую - никак. У нас нет власти над зависимостью y от X, коей, кстати, вообще может не быть. И нам нечего с этим поделать. Косвенно есть 2 способа: 1 - использовать другой ML-алгоритм, 2 - транформировать признаки. <br>

# 2. Метод наименьших квадратов OLS
Пусть
- X - дизайн-матрица размером $[n\times k]$,<br>
- $х_i$ - вектор из дизайн-матрицы размером $[k \times 1]$,<br>
- w - вектор коэффициентов размером $[k \times 1]$,<br>
- y - вектор респонсов размером $[n \times 1]$.<br>

Определим функцию потерь MSE как:

$$MSE(X, y, w) = \frac{\sum_{i=1}^n (\hat{y}(x_n) - y(x_n))^2}{n} = \frac{\sum_{i=1}^n (w^Tx_n-y(x_n))^2}{n} =  \frac{1}{n}((w^Tx_1 - y(x_1))^2 + (w^Tx_2 - y(x_2))^2 + \dots + (w^Tx_n - y(x_n))^2) $$

Сразу же бросается в глаза, что в правой части формула похожа на норму вектора-разности $Xw - y$, только без корня. И в с самом деле, это просто норма в квадрате. Таким образом,

$$MSE(X, y, w) = \frac{1}{n}\lVert\ Xw-y \rVert^{2}$$

Однако тут X и y - константы! Ведь по сути что такое < X, y >? - просто набор данных, который нам кто-то дал. Мы ничего не можем с ним поделать. Параметр, с которым мы можем "играть" чтобы в итоге получить наилучший fit (подгонку) - это **w**. Таким образом, можно записать:

$$MSE(w) = \frac{1}{n}\lVert\ Xw-y \rVert^{2} = \frac{1}{n}(Xw-y)^T(Xw-y) = \frac{1}{n}(w^TX^T-y^T)(Xw-y)$$

* <font color='brown'>Вопрос:</font> почему мы минимизируем именно MSE а не какую-либо другую функцию потерь, например MAE?
* <font color='brown'>Ответ:</font> можно минимизировать и MAE, ради бога. Тут две причины. Первая - с квадратами резидуалов работать проще, с точки зрения математики. Можешь попробовать взять mean absolute error в качестве функции потерь, и попробовать минизировать её; модуль доставит массу неприятностей. Если же брать не модули резидуалов, а их квадраты, то всё становится гораздо проще. Разумеется, можно было бы брать и другие степени резидуалов. Вторая причина - если мы минимизируем не MSE, то   тогда мы выйдем из условий [теоремы Маркова-Гаусса](https://en.wikipedia.org/wiki/Gauss%E2%80%93Markov_theorem), и наши эстиматоры $\hat{\beta_0}$ и $\hat{\beta_1}$ и перестанут быть лучшими среди линейных и несмещённых (BLUE - Best Linear Unbiased Estimator). Лучший несмещённый эстиматор - это тот, у которого наименьшая дисперсия среди множества линейных и несмещённых эстиматоров. <br>
*Почему мы выходим из условий теоремы Гаусса-Маркова?* Элементарно: эта теорема говорит о том, что при соблюдении её условий, OLS (МНК) является лучшим эстиматором среди линейных и несмещённых. OLS минимизирует именно MSE, так что если мы решим минимизировать какую-нибудь другую функцию потерь,то на полученный нами эстиматор уже не будут распространяться условия теоремы Гаусса-Маркова (ведь она говорит именно об OLS). <br>
Вместе с этим, есть и другие эстиматоры кроме OLS: есть например [LAD (Least Absolute Deviations)](https://en.wikipedia.org/wiki/Least_absolute_deviations) - эстиматор, который минимизирует не MSE, а SAE (sum of absolute errors, $\sum_{i=1}^n\lvert X\beta-y \rvert$). Он не имеет аналитического метода решения (простой формулы в явном виде). Если итоговый эстиматор в LAD получается линейным и несмещённым, то он будет хуже чем OLS, согласно теореме Гаусса-Маркова, если её условия соблюдаются. Есть ещё смещённые эстиматоры, например [James–Stein estimator](https://en.wikipedia.org/wiki/James%E2%80%93Stein_estimator).

<center><font color='grey'>**Теорема Гаусса-Маркова**</font></center>

Звучит следующим образом: при условии матожидания ошибок равного 0 (т.е. при строгой экзогенности) и сферичности ошибок (гомоскедастичность + отсутствие автокорреляции), оценки коэффициентов регрессии $\hat{\beta}$, полученные методом OLS, являются BLUE - Best Linear Unbiased Estimators, то есть лучшими среди линейный и несмещённых оценок. Лучшими - значит, имеющими наименьшую дисперсию среди любых других линейных и несмещённых оценок.

Предположения: предположения теоремы Гаусса-Маркова касаются ошибок (error terms).
1. Ошибки имеют матожидание 0: $\mathbb{E}[\varepsilon_i]=0$
2. Ошибки гомоскедастичны, то есть имеют одну и ту же конечную дисперсию: $Var[\varepsilon_i]=\sigma^2 < \infty$
3. Ошибки не коррелируют друг с дружкой: $Cov(\varepsilon_i, \varepsilon_j)=0 \quad \forall i \neq j$

Заметим, что теорема Гаусса-Маркова **не требует** Нормального распределения ошибок, или того, чтобы они были IID - nезависимыми и одинаково распределёнными. Нормальность ошибок может фигурировать как опциональное требование линейной регрессии, нужное для проведения статистических тестов; однако сама по себе теорема Гаусса-Маркова не требует нормальности ошибок ни в каком виде.

Сразу же рассмотрим Предположения, указанные выше, потому что они также являются Предположениями линейной регресии (конкретно - OLS).

><font color='red'>**Предположение 2:**</font> строгая экзогенность (strict exogeneity). "Матожидание ошибок равно нулю". Это предположение критически важно для теории МНК.
$$\mathbb{E}[\epsilon_i \mid x_1, x_2, \dots, x_n] = 0\space\space\space \forall i \in 1\dots{n}\space\space\space$$
<center>или просто</center><br>
$$\mathbb{E}[\epsilon \mid X] = 0$$
<font color='red'> Важность: критично</font>
<font color='red'>Объяснение:</font> предположение о строгой экзогенности переменных означает, что условное матожидание шума (ошибки, error) равно нулю. Условное - потому что оно рассматривается на заданной выборке X. <br>
Это предположение, в частности, означает, что среди независимых переменных Х отсутствуют так называемые "скрытые пременные" ([confounding variable](https://en.wikipedia.org/wiki/Confounding), lurking variable, משתנה מתערב), а также что независимые переменные не коррелируют с шумом: <br><br>
$$\mathbb{E}[X^T\epsilon]=0$$ <br>
Те регрессоры, которые коррелируют с шумом (ошибкой, error term), называются **эндогенными**, а те, которые не коррелируют с шумом - **экзогенными**. Так вот, строго _все_ независимые переменные должны экзогенными. <br>
Неплохой способ запомнить и рaзличать эти названия и их суть: экзо - это внешний, эндо - внутренний. Экзогенная переменная определена вне модели, то есть чтобы получить её значение, необходимо выйти "в мир" и реально провести требуемые замеры, иным способом достать значения экзогенной переменной не получится. Эндогенная же переменная определена внутри модели, то есть её значение можно тем или иным образом получить из других переменных. В этом разница между ними.<br><br>
[Приведём пример](https://stats.stackexchange.com/questions/218212/what-does-strict-exogeneity-condition-of-ols-really-mean/218222#218222). Респонс "Зарплата" генерируется "истинной" функцией <br><br>
$$y_i=\alpha+{\beta}x_i+{\gamma}z_i+u_i$$<br>
где $x$ - "Образование", $z$ - "Способности", $u$ - белый шум. Нам, однако, доступно только "Образование", а о существовании "Способностей" мы не знаем. Построив линейную регрессию зраплаты на образование, получим прогнозирующую функцию вида:<br><br>
$$ y_i =  \alpha + \beta x_i + \underbrace{\epsilon_i}_{\gamma z_i + u_i}$$
Как видим, в нашей модели error term включает в себя "Способности" и белый шум истинной функции, то есть $\epsilon=\gamma z_i + u_i$. Тогда $\mathbb{E}[\epsilon_i \mid x_i ] = \mathbb{E}[\gamma z_i + u_i \mid x_i] = \gamma\mathbb{E}[z_i \mid x_i]$ если $u_i$ - белый шум. Но если $\mathbb{E}[z_i \mid x_i] \neq 0$ и $\gamma \neq 0$, то $\mathbb{E}[\epsilon_i \mid x_i] \neq 0$ и предположение о строгой экзогенности нарушено!<br>
В данном случае это более чем вероятно, ведь "Способность" вряд ли будет равна нулю: в ВУЗы обычно идут более способные люди, а следовательно, практически наверняка будет $\gamma\neq 0$ и $\mathbb{E}[\epsilon_i \mid x_i] \neq 0$. Если бы мы регрессировали "Зарплату" на "Образование", то предположение о строгой экзогенности было бы нарушено.<br>
Имеем классический случай confounding variable: "Способности" *вызывают (cause! здесь causation как раз)* "Образование", а "Образование" влияет на "Зарплату". Подводя итог, видим, что в данном случае мы построили модель вида $y_i =  \alpha + \beta x_i+\epsilon_i$ и предполагаем строгую экзогенность, то есть $\mathbb{E}[\epsilon]=0$ , но на деле практически наверняка получаем $\mathbb{E}[\epsilon]\neq{0}$, что приведёт к неверным оценкам коэффициентов регресии.<br><br>
Строгая экзогенность - одно из условий теоремы Гаусса-Маркова, поэтому если это предположение не соблюдается, то мы выходим из теоремы.<br>
<font color='red'>Как проверить:</font> что в наших силах сделать, так это просто подсчитать среднее арифметическое резидуалов. Например, с помощью `np.mean(residuals)`. Полученное значение должно быть нулевым или быть крайне близко к нулю.<br>
<font color='red'>Нарушение:</font> что будет, если это предположение неверно? Оценки коэффициентов регрессии будут неверны. Смещение оценок коэффициентов регрессии называется **endogeneity bias** - смотри [прекрасное видео на тему](https://www.youtube.com/watch?v=HBr3376ttOg). В примере выше, получится что коэффициент $\hat{\beta}$ при $x$ ("Образовании") будет включать в себя как влияние Образования на Зарплату, так и влияние Способностей. Говоря точнее, $\hat{\beta}$ будет представлять собой влияние $x$ на $y$ плюс ассоциация $x$ и $z$ умноженная на влияние $z$ на $y$.
<font color='red'>Как исправить нарушение:</font> ? <br>
Также смотри короткое [видео](https://www.youtube.com/watch?v=a7_3qX67e7c) на эту тему.

<br><br>

><font color='red'>**Предположение 3:**</font> гетероскедастичность (heteroscedasticity). "Дисперсия ошибок постоянна". 
$$\mathbb{Var}[\varepsilon] = \sigma^2 < \infty$$
<font color='red'> Важность: достаточно важно </font>
<font color='red'>Объяснение:</font> данное предположение означает, что дисперсия ошибок на всех участках данных примерно постоянна.

<br><br>
Теперь продолжим работать с регрессией. Для минимизации функции потерь MSE найдём её градиент:

$$\frac{\partial}{\partial{w}}\frac{1}{n}(w^TX^T-y^T)(Xw-y) ;\quad \frac{\partial}{\partial{w}}w^TX^TXw - w^TX^Ty - y^TXw +y^Ty$$

Здесь мы убрали $1/n$ для удобства, ведь градиент будет приравниваться к нулю для поиска минимума, и $1/n$ там исчезнет, так что её можно не тащить через всё выведение формулы. Другой вариант: вообще-то OLS минимизирует не MSE, а RSS, а в ней вообще нет никакого $1/n$, там просто сумма квадратов резидуалов, так что всё даже проще.

Теперь идём по порядку:

1. $\frac{\partial}{\partial{w}}w^TX^TXw = 2w^T(X^TX)$ согласно правилу $\frac{\partial}{\partial{x}}x^TAx=2x^TA=2Ax$ для симметричных матриц A; а у нас в роли A выступает $X^TX$, которая несомненно симметрична, ведь $(X^TX)^T=X^TX$.
2. Рассмотрим $w^TX^Ty$ и $y^TXw$: по сути это одно и то же. В самом деле, транспонирование даёт нам 
$$(w^TX^Ty)^T = ((w^T)(X^Ty))^T = (X^Ty)^T(w) = y^TXw$$
то есть две эти матрицы - транспонированная версия друг друга. Однако они ещё и скаляры! В самом деле, осмотр размерностей показывает, что:
$$w^TX^Ty = [1\times k]\cdot [k\times n]\cdot [n\times 1] = [1\times n]\cdot [n\times 1]=[1\times{1}]=scalar=\alpha$$
Можно рассматривать этот скаляр как матрицу 1х1, содержащую только $\alpha$. Разумеется, сколько её ни транспонируй, а она останется неизменной: $[\alpha]^T=[\alpha]$. Таким образом, получаем, что
$$-w^TX^Ty-y^TXw = -2y^TXw$$
Воспользовавшись правилом $\frac{\partial}{\partial{x}}y^TAx = y^TA$ получим: 
$$\frac{\partial}{\partial{x}}(-2y^TXw)=-2y^TX$$
3. Рассмотрим $y^Ty$. Видим, что с точки зрения весов это вообще скаляр, так что его производная будет равна нулю.

<br><br>
Таким образом, имеем:

$$2w^TX^TX-2y^TX = 0 \implies w^TX^TX=y^TX \implies X^TXw = X^Ty \implies (X^TX)^{-1}X^TXw = (X^TX)^{-1}X^Ty \implies w = (X^TX)^{-1}X^Ty$$

В конечном итоге мы получили формулу для рассчёта коэффициентов регрессии

$$w = (X^TX)^{-1}X^Ty$$

## 4. Насколько близки подогнанные коэффициенты регрессии к истинным?

Здесь перед нами встают следующие две сложности:

1. <font color='green'>Ограниченность данных</font>. У нас нет доступа ко всей генеральной совокупности - у нас есть только выборка ограниченного размера. В силу этого обстоятельства есть дата-точки, которые не учтены нашей моделью регрессии, а значит, она вряд ли будет идеальной (т.е. точь-в-точь соответствующей истинной регрессии).<br>
*Что делать?* С самой проблемой - ничего: нам неоткуда добыть больше данных, чем те, которые у нас есть.
2. <font color='green'>Вариативность выборок (sampling variation)</font>. Для работы нам предоставлена конкретная выборка $\mathcal{D}_1$, и других выборок обычно нет в наличии. Однако если бы мы получили другую выборку $\mathcal{D}_2$, она состояла бы из других дата-точек и, соответственно, наша подгонка была бы другой, то есть - наши оценки коэффициентов $\hat{\beta_0}$ и $\hat{\beta_1}$ "истинной" регрессии были бы другими. Если бы нам дали совсем другую выборку $\mathcal{D}_3$, оценки коэффициентов $\hat{\beta_0}$ и $\hat{\beta_1}$ "истинной" регрессии были бы третьими. Это явление ведёт к тому, что имеющиеся у нас оценки обладают определённой долей **нестабильности**. <br>
*Что делать?* С самой проблемой - ничего: её решил бы только доступ к генеральной совокупности, а если бы у нас был к ней доступ, то надобность в регрессии и вообще в машинном обучении в принципе отпадала бы. Однако **можно оценить степень нестабильности оценок коэффициентов**: насколько сильно они зависят от конкретной предоставленной выборки? Насколько они статистически значимы? Каковы доверительные интервалы, в которых будет лежать истинное значение коэффициентов $\beta_0$ и $\beta_1$ с вероятностью 95% или 99%?

<br><br><br>
<center><font color='grey'>**Матожидание оценок коэффициентов регрессии β**</font></center>

Резонным будет вопрос, какое значение в среднем будут принимать наши оценки коэффициентов, если бы мы многократно брали бы выборки из популяции и рассчитывали на них регресиию? Иными словами, каково матожидание статистики "оценка коэффициента регрессии" $Е[\hat{\beta} \mid X]$?

Оказывается, при выполнении предположения о strict exogenity, наши оценки коэффициентов регрессии будут несмещёнными:

$$Е[\hat{\beta} \mid X]=\beta$$.
Vyvod:

<br><br><br>
<center><font color='grey'>**Стандартная ошибка оценок коэффициентов регрессии β (Standard Error)**</font></center>

**Стандартная ошибка статистики (standard error of a statistic)** - это стандартное отклонение выборочного распределения статистики. Обычно в роли интересующей нас статистики выступает эстиматор того или иного параметра, в данном случае - нас интересует стандартная ошибка эстиматора $\hat{\beta_i}$. <br>

*Принцип прост.* Если многократно вытягивать выборки из популяции и замерять на них интересующую нас статистику, то эта статистика будет иметь определённое распределение. Так вот, стандартное отклонение этого распределения - это и есть стандартная ошибка. <br>

*Что показывает стандартная ошибка?* Она является "мерой нестабильности оценки", то есть показывает, насколько наша оценка коэффициентов регрессии $\hat{\beta_i}$ зависит от конкретной выборки. Если $SE(\hat{\beta_i})$ небольшая, то можно сказать, что мы весьма уверены в нашей оценке $\beta_i$, тогда как если $SE(\hat{\beta_i})$ велика, то мы не уверены в оценке $\beta_i$. <br>
Это логично. Большое значение $SE(\hat{\beta_i})$ означает, что наша оценка $\hat{\beta_i}$ сильно варьируется в зависимости от конкретной выборки, а значит, мы не уверены в правильности $\hat{\beta_i}$. И наоборот.

Стандартные ошибки оценок коэффициентов рассчитываются по формулам:

$$\hat{SE}(\hat{\beta_0})^2 = \hat{\sigma^2}(\epsilon) \bigg[ \frac{1}{n} + \frac{\bar{x}^2}{\sum_{i=1}^n (x_i - \bar{x})^2} \bigg] , \space\space\space \hat{SE}(\hat{\beta_1})^2 = \frac{\hat{\sigma^2}(\epsilon)}{\sum_{i=1}^n (x_i - \bar{x})^2}$$

где $\hat{\sigma^2}$ или иначе говоря $\hat{Var}(\epsilon)$ - дисперсия вектора резидуалов. Иначе это называется **стандартная ошибка резидуала (RSE - Residual Standard Error)**. В конце концов, выборочный резидуал $\hat{\epsilon}$ - это тоже статистика на выборке, как и $\hat{\beta}$, которая может быть использована в качестве эстиматора для резидуала популяции - то есть параметра популяции $\epsilon$. Поскольку параметр "резидуал" только оценивается а не рассчитывается (ведь у нас есть только выборка, а не вся популяция), то над стандартными ошибками стоит "шапка": $\hat{SE(\beta)}$ у нас - тоже эстимейты, оценки.<br>
Оценка стандартной ошибки резидуала выполняется по формуле:

$$RSE = \frac{\sqrt{RSS}}{n-2} = \frac{\sqrt{\sum_{i=1}^{n}(\hat{y}_i-y_i)^2}}{n-2} = \frac{\sqrt{e_1^2 + e_2^2 + \dots + e_n^2}}{n-2}=\frac{\lvert\lvert e \rvert\rvert}{n-2}$$

```python
# Для β1 - при рассчёте используется формула:
# (np.sqrt(((np.linalg.norm(residuals))**2) / (SAMPLE_SIZE-2))) / (np.linalg.norm(X_sample-np.mean(X_sample)))

print('Оценки коэффициентов регрессии (напомним - сделаны на выборке):\n\t· Оценка β0:',
      statsmodels_result.params[0], '\n\t· Оценка β1:', statsmodels_result.params[1])
print('Оценки стандартных ошибок оценок коэффициентов:\n\t· Оценка SE(β0)', statsmodels_result.bse[0],
     '\n\t· Оценка SE(β0)', statsmodels_result.bse[1])
```

Видим, что оценки $\hat{SE}(\hat{\beta_i})$ небольшие, поэтому мы достаточно уверены в рассчитаных нами оценках истинных коэффициентов регрессии. 

> ... предположение о гомоскедастичности?

Но насколько именно оценки стандартных ошибок $\hat{SE}(\hat{\beta}$) должны быть малы, чтобы мы могли смело утверждать, что рассчитанные нами оценки коэффициентов достаточно стабильны?

Допустим, мы регрессируем "Зарплату" на "Образование". Мы получили результат: $\hat{\beta_1}=1$. Нас интересует стабильность оценки $\hat{\beta_1}$, то есть наклона прямой регрессии в данном случае. Находится ли полученное значение 1 достаточно далеко от нуля? Или, напротив, лежит очень близко к нулю, а значит, весьма вероятно что оно вызвано выборочной вариативностью и не отражает реальной зависимости между з/п и образованием?

Рассчитаем стандартную ошибку наклона $SE(\hat{\beta_1})$.
* Если $SE(\hat{\beta_1})$ = 0.2, то мы видим, что $\hat{\beta_1}=1$ находится на удалении в 5 стандартных ошибок (стандартных отклонений) от нуля, а значит, вероятность того, что полученное значение $\hat{\beta_1}$ незначительно - крайне мала. Таким образом, мы делаем вывод, что почти наверняка существует связь между образованием и зарплатой.
* Если $SE(\hat{\beta_1})$ = 2, то $\hat{\beta_1}=1$ находится на удалении всего в $1/2$ стандартной ошибки от нуля, то есть скорее всего нет связи между образованием и зарплатой.
<br><br>

> Примечание: гетероскедастичность уводит в сторону оценки стандартных ошибок, доверительных интервалов, t-тестов и p-values, то есть делает их смещёнными, а значит, неверными. Соответственно, наша модель будет систематически делать неверные выводы о статистической значимости оценок коэффициентов, систематически завышая или занижая результаты этих тестов и стандартных ошибок, а также будет выдавать неверные доверительные интервалы для $\hat{\beta}$ - то есть такие интервалы, которые будут иметь неверное покрытие из-за того, что они смещены вместе со стандартной ошибкой.<br> Впрочем, гетероскедастичность будет мешать только когда она достаточно велика. Если она не слишком большая, то тесты отработают достаточно хорошо.

# Вывод формул

## 1. Вывод наклона линии регрессии

Вывод находится [здесь](http://www.alfisol.com/IFS/IFS-003/DerivingLinearRegressions.php) вместе с интересной статьёй. При рассчёте параметров прямой, имеющей наименьшую сумму квадратов резидуалов RSS, можно было бы просто перебирать различные варианты прямых, рассчитывать их RSS, сравнивать, и в итоге отобрать одну, у которой RSS минимальна. С компьютером эта процедура выполняется очень просто, но в 1700х не было компьютеров, поэтому в 1975 году Карл Фридрих Гаусс, которому тогда было 17 лет, разработал метод линейной регресии. Вместо того, чтобы перебором искать линию, минимизирующую RSS, мы решим задачу "вспять": минимизируем RSS, и уже оттуда вычислим параметры линии, которая дала нам минимальную RSS; отсюда и термин "регрессия". По другому мнению, этот термин означает регрессию к среднему; например, регрессию роста сыновей высоких или низких родителей обратно к среднему росту населения.

Итак, имеется функция $y=\hat{\beta_0}+\hat{\beta_1}x_1$, а также $RSS = \sum(y_i - \hat{y_i})^2$. Распишем вторую формулу подробнее и раскроем квадрат:

$$RSS=\sum(y_i-\hat{\beta_0}-\hat{\beta_1}x_i)^2=\sum(y^2_i-2y_i\hat{\beta_0}-2y_i\hat{\beta_1}x_i+2\hat{\beta_0}\hat{\beta_1}x_i+\hat{\beta_1^2}x_i^2+\hat{\beta^2_0})$$

Формула выше представляет собой график RSS как функции от оценок коэффициентов линейной регресии: $RSS(\hat{\beta_0}, \hat{\beta_1})$. Найдём минимум этой функции, взяв производные по коэффициентам:

$$\frac{\partial}{\partial \hat{\beta_0}}RSS=-2\sum{y_i}+2\hat{\beta_1}\sum{x_i}+2\sum{\hat{\beta_0}}$$
$$\frac{\partial}{\partial \hat{\beta_1}}RSS=2\hat{\beta_1}\sum{x^2_i}+2\hat{\beta_0}\sum{x_i}-2\sum{y_ix_i}$$

Для упрощения нотации запишем: пусть $S_X=\sum{x_i}, S_Y=\sum{y_i}, S_{YX}=\sum{y_ix_i}, S_{XX}=\sum{x^2_i}, nb_0=\sum{\hat{\beta_0}}$. Тогда:

$$\frac{\partial}{\partial \hat{\beta_0}}RSS=0=-2S_{Y}+2\hat{\beta_1}S_X+2nb_0 \quad\quad ; \quad\quad \frac{\partial}{\partial \hat{\beta_1}}RSS=0=2\hat{\beta_1}S_{XX}+2\hat{\beta_0}S_X - 2S_{YX}$$

Отсюда

$$\hat{\beta_0}=b_0=\frac{S_Y-\hat{\beta_1}S_X}{n} \implies S_Y=\hat{\beta_1}S_X+nb_0 \quad\quad; \quad\quad \hat{\beta_1}=\frac{S_{YX} - \hat{\beta_0}S_X}{S_{XX}} \implies S_{YX} = \hat{\beta_1}S_{XX} + \hat{\beta_0}S_X $$

Подставив выражение для $\hat{\beta_0}$ в выражение для  $\hat{\beta_1}$, получим:

$$0=\hat{\beta_1}S_{XX}+S_X\frac{S_Y-\hat{\beta_1}S_X}{n}-S_{YX}$$
$$0=\hat{\beta_1}S_{XX}+\frac{S_XS_Y}{n}-\frac{\hat{\beta_1}S_XS_X}{n}-S_{YX}$$
$$0=n\hat{\beta_1}S_{XX}+S_XS_Y-\hat{\beta_1}S_XS_X-nS_{YX}$$
$$\hat{\beta_1}(nS_{XX}-S_XS_X) = nS_{YX}-S_XS_Y$$
<br>
$$\hat{\beta_1} = \frac{nS_{YX}-S_XS_Y}{nS_{XX}-S_XS_X} = \frac{n\sum{y_ix_i} - \sum{x_i}\sum{y_i}}{n\sum{x^2_i}-(\sum{x_i})^2}$$

Рассмотрим числитель и знаменатель отдельно. Начнём со знаменателя, воспользовавшись формулой "Var[Х] = mean of square minus square of mean":

$$n\sum{x^2_i}-(\sum{x_i})^2=n\cdot{n}\cdot \underbrace{ \frac{1}{n}\sum{x^2_i} }_{\mathrm{E}[X^2]} - n\cdot \underbrace{ \frac{1}{n}\sum{x_i} }_{\mathrm{E}[X]} \cdot n\cdot \underbrace{ \frac{1}{n}\sum{x_i} }_{\mathrm{E}[X]} = n^2(\mathrm{E}[X^2] - (\mathrm{E}[X])^2) = n^2 \cdot \mathrm{Var}(X)$$

Что можно расписать как

$$n^2 \cdot \mathrm{Var}(X)=n^2 \cdot \frac{1}{n}\sum(x_i-\mathrm{E}[X])^2 = n \cdot \sum{(x_i-\bar{x})^2}$$

Теперь рассмотрим числитель. Там имеем:
$$n\sum{y_ix_i} = n \cdot n\cdot \underbrace{ \frac{1}{n}\sum{x_iy_i} }_{\mathrm{E}[XY]} = n^2\cdot\mathrm{E}[XY]$$
$$\sum{x_i}\sum{y_i} = n\cdot\underbrace{ \frac{1}{n}\sum{x_i} }_{\mathrm{E}[X]} \cdot n\cdot \underbrace{ \frac{1}{n}\sum{y_i} }_{\mathrm{E}[Y]}=n^2\cdot\mathrm{E}[X]\mathrm{E}[Y]$$

Итого в числителе получим:

$$n\sum{y_ix_i} - \sum{x_i}\sum{y_i}=n^2\cdot\mathrm{E}[XY] - n^2\cdot\mathrm{E}[X]\mathrm{E}[Y] = n^2\cdot(\mathrm{E}[XY] - \mathrm{E}[X]\mathrm{E}[Y])$$

Сразу бросается в глаза формула ковариации $Cov(X, Y) = \mathrm{E}[XY] - \mathrm{E}[X]\mathrm{E}[Y] = \frac{1}{n}\sum(x_i-\bar{x})(y_i-\bar{y})$, воспользуемся ей:

$$n^2\cdot \underbrace{ (\mathrm{E}[XY] - \mathrm{E}[X]\mathrm{E}[Y]) }_{\mathrm{Cov}(X, Y)} = n^2 \cdot \frac{1}{n}\sum(x_i-\bar{x})(y_i-\bar{y}) = n\cdot\sum(x_i-\bar{x})(y_i-\bar{y}) $$

Итого получим:
<font color='maroon'>
$$\hat{\beta_1} = \frac{n\cdot\sum(x_i-\bar{x})(y_i-\bar{y})}{n \cdot \sum{(x_i-\bar{x})^2}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum{(x_i-\bar{x})^2}}$$
</font>

Собственно, эту формулу мы и видим в книге ISLR.

Нетрудно заметить из выкладок выше, что данную формулу можно также записать как:
<br><br>
<font color='maroon'>
$$\hat{\beta_1} = \frac{\mathrm{Cov}(XY)}{\mathrm{Var}[X]}$$
</font>

<br><br><br>
## 2. Несмещённость оценок коэффициентов регрессии

Докажем несмещённость оценки наклона линии регрессии. Начнём с наклона прямой. Вспомним, что формула для него может быть записана следующим образом:

$$\hat{\beta_1} = \frac{n\sum{y_ix_i} - \sum{x_i}\sum{y_i}}{n\sum{x^2_i}-(\sum{x_i})^2} = \frac{ n\sum{y_ix_i} - n\cdot\overbrace{ \frac{1}{n}\cdot\sum{x_i} }^{среднее}\cdot{n}\cdot\overbrace{ \frac{1}{n}\cdot\sum{y_i} }^{среднее} }{n\sum{x_i^2}-n\cdot\underbrace{ \frac{1}{n}\cdot\sum{x_i} }_{среднее}\cdot{n}\cdot\underbrace{ \frac{1}{n}\cdot\sum{x_i} }_{среднее} }=\frac{n\sum{y_ix_i} - n^2\bar{x}\bar{y}}{n\sum{x_i^2}-n^2\bar{x}^2} =\frac{\sum{y_ix_i} - n\bar{x}\bar{y}}{\sum{x_i^2}-n\bar{x}^2}$$

Также обратим внимание, что числитель может быть переписан следующим образом:

$$\hat{\beta_1} = \frac{\sum{y_ix_i} - n\bar{x}\bar{y}}{\sum{x_i^2}-n\bar{x}^2} = \frac{\sum{y_ix_i} - \bar{x}\sum{y_i}}{\sum{x_i^2}-n\bar{x}^2}=\frac{\sum({x_i-\bar{x}})y_i}{\sum{x_i^2}-n\bar{x}^2}$$

Рассмотрим условное матожидание:

$$\mathrm{E}[\hat{\beta_1} \mid x_i] = \mathrm{E}\Bigg[ \frac{\sum({x_i-\bar{x}})y_i}{\sum{x_i^2}-n\bar{x}^2} \Bigg] = \frac{1}{\sum{x_i^2}-n\bar{x}^2}\sum({x_i-\bar{x}})\mathrm{E}[y_i \mid x_i]$$

Однако по гипотезе линейности, $E[y_i \mid x_i] = E[\beta_0+\beta_1x_i+\mu\mid x_i]=\beta_0+\beta_1x_i$. Здесь мы использовали линейность матожидания и <font color='green'>Предположение о строгой экзогенности</font>, т.е. что $\mathrm{E}[\mu \mid x_i]=0$. Отсюда:

$$\mathrm{E}[\hat{\beta_1} \mid x_i] = \frac{\sum(x-\bar{x})(\beta_0+\beta_1x_i)}{\sum{x_i^2}-n\bar{x}^2}= \frac{ \sum (x_i\beta_0+x_i^2\beta_1-\bar{x}\beta_0-\bar{x}x_i\beta_1) }{\sum{x_i^2}-n\bar{x}^2}=\frac{ \beta_0\overbrace{\sum x_i}^{=n\bar{x}}-\beta_0\overbrace{\sum\bar{x}}^{=n\bar{x}}+\beta_1\sum\left( x_i^2-x_i\bar{x}\right) }{\sum{x_i^2}-n\bar{x}^2}=\frac{\beta_1\sum({x_i^2-x_i\bar{x}})}{\sum{x_i^2}-n\bar{x}^2}$$

Легко заметить, что

$$\mathrm{E}[\hat{\beta_1}\mid x_i] = \frac{\beta_1(\sum{x_i^2}-\bar{x}\sum{x_i})}{\sum{x_i^2}-n\bar{x}^2}=\frac{\beta_1(\sum{x_i^2}-\bar{x}\cdot{n}\cdot \overbrace{ \frac{1}{n}\sum{x_i} }^{\bar{x}})}{\sum{x_i^2}-n\bar{x}^2} = \frac{\beta_1(\sum{x_i^2}-n\bar{x}^2)}{(\sum{x_i^2}-n\bar{x}^2)}=\beta_1$$

Отсюда, воспользовавшись [Law of total expectation](https://en.wikipedia.org/wiki/Law_of_total_expectation), получим наконец, что
<br><br>

<font color='maroon'>
$$\mathrm{E}[\hat{\beta_1}] = \beta_1$$
</font>

Заметим, что в доказательстве мы использовали:
1. Предположение 1 о линейной зависимости: $\mathrm{E}[y_i\mid x_i] = \mathrm{E}[\beta_0+\beta_1x_i+\mu\mid x_i]=\beta_0+\beta_1x_i$
2. Предположение 2 о строгой экзогенности: $\mathrm{E}[\mu \mid x_i]=0$
3. Дата-точки $(x_i, y_i)$ - iid (??? а надо ли это? Если теорема Гаусса-Маркова этого не требует, то сие тем более не должно!...) С другой стороны, теорема Гаусса-Маркова не требует iid ошибок errors, а тут говорится об iid **дата-точек**.
